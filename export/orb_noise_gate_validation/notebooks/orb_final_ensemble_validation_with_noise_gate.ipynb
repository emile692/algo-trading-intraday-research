# ORB Final Ensemble Validation With Noise Gate

Validation ciblee de l'overlay noise-area gate sur baseline figee.


```python
from pathlib import Path
import pandas as pd
from IPython.display import display, Image, Markdown

root = Path.cwd()
out = root / 'export' / 'orb_noise_gate_validation'
print(out)
```

```python
results = pd.read_csv(out / 'noise_gate_validation_results.csv')
top = pd.read_csv(out / 'noise_gate_top_configs.csv')
summary = (out / 'noise_gate_summary.md').read_text(encoding='utf-8')

display(top[['name','run_group','noise_style','noise_lookback','noise_vm','noise_k','confirm_bars','schedule','ranking_prop_score','oos_net_pnl','oos_max_drawdown','oos_worst_5day_drawdown','oos_max_losing_streak']].head(15))
```

```python
baseline = results[results['name']=='baseline_fixed'].iloc[0]
noise = results[(results['noise_gate_enabled']==True) & (results['status']=='ok')].sort_values('ranking_prop_score', ascending=False)
compare = pd.concat([pd.DataFrame([baseline]), noise.head(8)], ignore_index=True)
display(compare[['name','run_group','noise_style','noise_lookback','noise_vm','noise_k','confirm_bars','schedule','ranking_prop_score','oos_net_pnl','oos_sharpe_ratio','oos_profit_factor','oos_expectancy','oos_max_drawdown','oos_return_over_drawdown','oos_worst_day','oos_worst_5day_drawdown','oos_max_losing_streak','oos_nb_trades','oos_pct_days_traded']])
```

```python
charts = out / 'charts'
for name in [
    'heatmap_prop_score_L_VM.png',
    'heatmap_maxdd_L_VM.png',
    'heatmap_netpnl_L_VM.png',
    'heatmap_L_k.png',
    'heatmap_confirm_schedule.png',
    'heatmap_synthesis_best_by_family.png',
    'equity_curves_top_noise_gate.png',
    'drawdown_top_noise_gate.png',
]:
    p = charts / name
    if p.exists():
        display(Image(filename=str(p)))
```
